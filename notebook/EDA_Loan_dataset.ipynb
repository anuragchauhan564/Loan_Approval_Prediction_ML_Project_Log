# Loan Approval Prediction

# EDA on Loan Dataset

## About this file
### The loan approval dataset is a collection of financial records and associated information used to determine the eligibility of individuals or organizations for obtaining loans from a lending institution. 
#### It includes various factors such as cibil score, income, employment status, loan term, loan amount, assets value, and loan status. This dataset is commonly used in machine learning and data analysis to develop models and algorithms that predict the likelihood of loan approval based on the given features.

### Data Visualization

```python
# percentage for pie chart
```

```python
 df['education'].value_counts()
```

```python
education_percentage = df['education'].value_counts(normalize=True) * 100  #  percentage for Graduate and Not Graduate
```

```python
education_lables = ['Graduate','Not Graduate']
```

```python
plt.figure(figsize=(5,5))
plt.pie(education_percentage, labels=  education_lables,autopct='%.0f%%')
plt.title("Education Class Distribution")
plt.show()
```

```python
self_employed_percentage = df['self_employed'].value_counts(normalize=True)*100 # percentage for Yes and No in self_employed
```

```python
self_employed_lables = ['Yes','No']
```

```python
plt.figure(figsize=(5,5))
plt.pie(self_employed_percentage, labels=  self_employed_lables,autopct='%1.01f%%')
plt.title("Self Employed Class Distribution")
plt.show()
```

```python
percentage = df['loan_status'].value_counts(normalize=True) * 100 # percentage for  'Approved','Rejected' in loan_status 
```

```python
loan_lables = ['Approved','Rejected']
```

```python
plt.figure(figsize=(5,5))
plt.pie(percentage, labels=  loan_lables,autopct='%1.01f%%')
plt.title("Loan Status Class Distribution")
plt.show()
```
